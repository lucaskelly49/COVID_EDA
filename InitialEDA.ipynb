### Meta Data:

https://www.kaggle.com/rohanrao/covid19-forecasting-metadata

```python
df1 = pd.read_csv('region_metadata.csv')
```

```python
df2 = pd.read_csv('region_date_metadata.csv')
```

```python
df3 = pd.merge(df1, df2, on='Country_Region')
```

```python
df_meta_us = df3.loc[df3['Country_Region']=='US']
```

```python
df_meta_us.drop(['Date', 'Province_State_y', 'Recoveries'], axis=1, inplace=True)
```

```python
df_meta_us = df_meta_us.rename(columns={"Province_State_x": "Province_State"})
```

### Merged
